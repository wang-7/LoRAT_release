```python
import numpy as np
import matplotlib.pyplot as plt

finetuned = [19.62, 17.59, 16.44, 24.60, 16.96]
globalmodel = [19.03, 15.73, 15.85, 23.37, 15.33]

try:
    if len(finetuned) != len(globalmodel):
        raise ValueError("The lengths of 'finetuned' and 'globalmodel' arrays must be the same.")
except ValueError as e:
    print(f"Error: {e}")
    raise

x = np.arange(len(finetuned))
width = 0.35

plt.rcParams['pdf.fonttype'] = 42  
plt.figure(figsize=(10, 6))
plt.bar(x - width/2, finetuned, width, label='Adapted Model', color='#8da0cb')
plt.bar(x + width/2, globalmodel, width, label='Base Model', color='#66c2a5')

plt.ylabel('Average SNR', fontsize=16)
plt.xlabel('Base Station', fontsize=16)
plt.xticks(x, ['BS 1', 'BS 2', 'BS 3', 'BS 4', 'BS 5'], fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize=14)
plt.savefig('finetuned_comparison.pdf', bbox_inches='tight', format='pdf', dpi=300)
plt.show()
```

```python
# lora r
different_r = [17.45, 17.71, 17.83, 17.98, 18.05]
full = 18.01
# plt.figure(figsize=(10, 6))
plt.figure(figsize=(8, 6))
plt.bar(range(1, 6), different_r, width=0.7, color='#8da0cb')
plt.axhline(y=full, color='darkred', linestyle='--', label='Full-Parameter Fine-tuned')
plt.xlabel('LoRA Rank')
plt.ylabel('Average SNR')
plt.legend()
plt.xticks(range(1, 6), ['1', '2', '4', '8', '16'])
plt.ylim(16, 19)
plt.savefig('lora_r.pdf', bbox_inches='tight', format='pdf', dpi=300)
plt.show()
```
